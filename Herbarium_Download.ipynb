```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
import csv
import os
import pandas
import csv
import requests
from PIL import Image
import threading
import time
from threading import Thread
from _thread import start_new_thread, allocate_lock
import shutil
from PIL import ImageFile
```

```python
AS =pandas.read_pickle("idigbio_accepted_salix_bebbiana.pkl")
AS.drop_duplicates(subset = ['accessuri'],keep = 'first', inplace = True)
AS['institutioncode'] = AS['institutioncode'].str.replace('\W', '')
AS['catalognumber'] = AS['catalognumber'].str.replace('\W', '')
AS['Sample_id'] = AS['institutioncode'].astype(str) + '_' + AS['catalognumber'].astype(str)
AS['Image_id'] = AS.Sample_id + '_' + (AS.groupby('Sample_id').cumcount()+1).astype(str)
AS.reset_index(level = 0, inplace = True)
AS.drop(['index'],axis = 1,inplace = True)
```

```python
sum(pandas.isnull(AS['accessuri']))
AS = AS.dropna(axis=0, subset=['accessuri'])
```

```python
for index, row in AS.iterrows():
    Spieces_name= row.scientificname
    parent_folder = os.path.join('./Images/', Spieces_name)
    if not os.path.exists(parent_folder):
        os.mkdir(parent_folder)
        
    file= row.Image_id
    url = row.accessuri
    print(url)
    result = requests.get(url, stream=True)
    time.sleep(4)
    if result.status_code == 200:
        child_folder = os.path.join(parent_folder, file)
        if not os.path.exists(child_folder):
            os.mkdir(child_folder)
        full_path = os.path.join(child_folder, '{}.jpg'.format(file))
        with open(full_path, 'wb') as f:
            result.raw.decode_content = True
            shutil.copyfileobj(result.raw, f)
        
        try:
            image =  Image.open(full_path)
            image = image.resize((1900, 2400))
            ImageFile.LOAD_TRUNCATED_IMAGES = True
            image.save(full_path)
        except:
            pass

        
print('done')
```
